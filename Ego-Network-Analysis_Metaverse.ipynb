```python
def scroll(modal):
    try:        
        # Ïä§ÌÅ¨Î°§ ÎÜíÏù¥ Î∞õÏïÑÏò§Í∏∞
        last_height = driver.execute_script("return arguments[0].scrollHeight", modal)
        while True:
            pause_time = random.uniform(0.5, 0.8)
            # ÏµúÌïòÎã®ÍπåÏßÄ Ïä§ÌÅ¨Î°§
            driver.execute_script("arguments[0].scrollTo(0, arguments[0].scrollHeight);", modal)
            # ÌéòÏù¥ÏßÄ Î°úÎî© ÎåÄÍ∏∞
            time.sleep(pause_time)
            # Î¨¥Ìïú Ïä§ÌÅ¨Î°§ ÎèôÏûëÏùÑ ÏúÑÌï¥ ÏÇ¥Ïßù ÏúÑÎ°ú Ïä§ÌÅ¨Î°§
            driver.execute_script("arguments[0].scrollTo(0, arguments[0].scrollHeight-50);", modal)
            time.sleep(pause_time)
            # Ïä§ÌÅ¨Î°§ ÎÜíÏù¥ ÏÉàÎ°≠Í≤å Î∞õÏïÑÏò§Í∏∞
            new_height = driver.execute_script("return arguments[0].scrollHeight", modal)
            try:
                # 'ÎçîÎ≥¥Í∏∞' Î≤ÑÌäº ÏûàÏùÑ Í≤ΩÏö∞ ÌÅ¥Î¶≠
                all_review_button = driver.find_element_by_xpath('/html/body/div[1]/div[4]/c-wiz/div/div[2]/div/div/main/div/div[1]/div[2]/div[2]/div/span/span').click()
            except:
                # Ïä§ÌÅ¨Î°§ ÏôÑÎ£å Í≤ΩÏö∞
                if new_height == last_height:
                    print("Ïä§ÌÅ¨Î°§ ÏôÑÎ£å")
                    break
                last_height = new_height
                
    except Exception as e:
        print("ÏóêÎü¨ Î∞úÏÉù: ", e)
```

```python
# Î¶¨Î∑∞ Îç∞Ïù¥ÌÑ∞ ÌÅ¥ÎûòÏä§ Ï†ëÍ∑º
review_source = soup_source.find_all(class_ = 'RHo1pe')
# Î¶¨Î∑∞ Îç∞Ïù¥ÌÑ∞ Ï†ÄÏû•Ïö© Î∞∞Ïó¥
dataset = []
# Îç∞Ïù¥ÌÑ∞ ÎÑòÎ≤ÑÎßÅÏùÑ ÏúÑÌïú Î≥ÄÏàò
review_num = 0 
# Î¶¨Î∑∞ 1Í∞úÏî© Ï†ëÍ∑ºÌï¥ Ï†ïÎ≥¥ Ï∂îÏ∂ú
for review in tqdm(review_source):
    review_num+=1
    # Î¶¨Î∑∞ Îì±Î°ùÏùº Îç∞Ïù¥ÌÑ∞ Ï∂îÏ∂ú
    date_txt   = review.find_all(class_ = 'bp9Aid')[0].text
    date_full  = datetime.strptime(date_txt, '%B %d, %Y').strftime('%Y-%m-%d')
    date_year  = datetime.strptime(date_txt, '%B %d, %Y').strftime('%Y') # Ïó∞ÎèÑ Îç∞Ïù¥ÌÑ∞ Ï∂îÏ∂ú
    date_month = datetime.strptime(date_txt, '%B %d, %Y').strftime('%m') # Ïõî Îç∞Ïù¥ÌÑ∞ Ï∂îÏ∂ú
    date_day   = datetime.strptime(date_txt, '%B %d, %Y').strftime('%d') # Ïùº Îç∞Ïù¥ÌÑ∞ Ï∂îÏ∂ú

    user_name = review.find_all(class_ = 'X5PpBb')[0].text # ÎãâÎÑ§ÏûÑ Îç∞Ïù¥ÌÑ∞ Ï∂îÏ∂ú
    rating = review.find_all(class_ = "iXRFPc")[0]['aria-label'][6] # ÌèâÏ†ê Îç∞Ïù¥ÌÑ∞ Ï∂îÏ∂ú
    content = review.find_all(class_ = 'h3YV2d') # Î¶¨Î∑∞ Îç∞Ïù¥ÌÑ∞ Ï∂îÏ∂ú

    data = {
        "id"       : review_num, 
        "date"     : date_full,
        "date_year" : date_year,
        "date_month": date_month,
        "date_day"  : date_day,
        "rating"   : rating,
        "user_name" : user_name,
        "content"  : content
    }
    dataset.append(data)
```

```python
def compute_coherence_values(dictionary, corpus, texts, limit, start=2, step=3):
    coherence_values = []
    model_list = []
    
    for num_topics in range(start, limit, step): 
        model = gensim.models.ldamodel.LdaModel(corpus=corpus, id2word=id2word, num_topics=num_topics) 
        model_list.append(model) 
        coherencemodel = CoherenceModel(model=model, texts=texts, dictionary=dictionary, coherence='c_v') 
        coherence_values.append(coherencemodel.get_coherence())        
    
    return model_list, coherence_values

def find_optimal_number_of_topics(dictionary, corpus, processed_data): 
    limit = 8;
    start = 2; 
    step = 2; 
    
    model_list, coherence_values = compute_coherence_values(dictionary=dictionary, corpus=corpus, texts=processed_data, start=start, limit=limit, step=step) 
    x = range(start, limit, step) 
    
    plt.plot(x, coherence_values) 
    plt.xlabel("Num Topics")
    plt.ylabel("Coherence score") 
    plt.legend(("coherence_values"), loc='best') 
    plt.show()
```

```python
# -*- encoding: utf-8 -*-
import pandas as pd
from tqdm import tqdm

if __name__ == '__main__':
    # Îã®Ïñ¥ÏåçÏùò ÎπàÎèÑÎ•º Ï≤¥ÌÅ¨ÌïòÍ∏∞ÏúÑÌï¥ DTMÏùÑ Î∂àÎü¨Ïò®Îã§.
    dataset = pd.read_csv(r'C:\Users\qorrl\Desktop\raw_data\DTM\zepeto_review_dataset_neg_DTM.csv')
    # Îã®Ïñ¥Îì§Ïùò Î™©Î°ùÏùÑ Í∞ÄÏ†∏Ïò®Îã§.
    # Ïù¥Îïå 0Î≤àÏß∏ Ïù∏Îç±Ïä§ÏóêÎäî Îπà Ïπ∏Ïù¥ Îì§Ïñ¥Ïò§ÎØÄÎ°ú Ïù∏Îç±Ïã±ÏùÑ ÌÜµÌï¥ ÏóÜÏï†Ï§ÄÎã§.
    column_list = dataset.columns[1:]
    word_length = len(column_list)

    # Í∞Å Îã®Ïñ¥ÏåçÏùò ÎπàÎèÑÏàòÎ•º Ï†ÄÏû•Ìï† dictionary ÏÉùÏÑ±
    count_dict = {}

    for doc_number in tqdm(range(len(dataset)), desc='Îã®Ïñ¥Ïåç ÎßåÎì§Í∏∞ ÏßÑÌñâÏ§ë'):
        tmp = dataset.loc[doc_number]           # ÌòÑÏû¨ Î¨∏ÏÑúÏùò Îã®Ïñ¥ Ï∂úÌòÑ ÎπàÎèÑ Îç∞Ïù¥ÌÑ∞Î•º Í∞ÄÏ†∏Ïò®Îã§.
        for i, word1 in enumerate(column_list):
            if tmp[word1]:              # ÌòÑÏû¨ Î¨∏ÏÑúÏóê Ï≤´Î≤àÏß∏ Îã®Ïñ¥Í∞Ä Ï°¥Ïû¨Ìï† Í≤ΩÏö∞
                for j in range(i + 1, word_length):
                    if tmp[column_list[j]]:              # ÌòÑÏû¨ Î¨∏ÏÑúÏóê ÎëêÎ≤àÏß∏ Îã®Ïñ¥Í∞Ä Ï°¥Ïû¨Ìï† Í≤ΩÏö∞
                        count_dict[column_list[i], column_list[j]] = count_dict.get((column_list[i], column_list[j]), 0) + max(tmp[word1], tmp[column_list[j]])

    # count_listÏóê word1, word2, frequency ÌòïÌÉúÎ°ú Ï†ÄÏû•Ìï† Í≤ÉÏù¥Îã§.
    count_list = []

    for words in count_dict:
        count_list.append([words[0], words[1], count_dict[words]])

    # Îã®Ïñ¥Ïåç ÎèôÏãú Ï∂úÌòÑ ÎπàÎèÑÎ•º DataFrame ÌòïÏãùÏúºÎ°ú ÎßåÎì†Îã§.
    df = pd.DataFrame(count_list, columns=["word1", "word2", "freq"])
    df = df.sort_values(by=['freq'], ascending=False)
    df = df.reset_index(drop=True)

    df.to_csv(r'C:\Users\qorrl\Desktop\raw_data\DTM\zepeto_review_dataset_neg_DTM_fre.csv', encoding='utf-8-sig')
```
